```python
import pandas as pd
```

```python
answer = pd.read_csv('../dataset/add_data.csv').head(20)
response = pd.read_csv('responses/response.csv').head(20)
```

```python
# for column in answer.columns:
compare = []
for i in range(2, 37):
    compare.append(answer[f'answer_{{{i}}}'].compare(response[f'answer_{{{i}}}']))
```

```python
for i in range(len(compare)):
    print(f"base {i+2}: {compare[i].shape[0]}")
```

```python
for i in range(len(compare)):
    display(compare[i].rename(columns={"self": f'answer_{{{i+2}}}', "other": f'response_{{{i+2}}}'}))
```
